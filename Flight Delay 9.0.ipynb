```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
%matplotlib inline
```

```python
data = pd.read_csv('Train.csv')
```

```python
data.head()
```

```python
data = data[['ID','target','DATOP','FLTID','DEPSTN','ARRSTN','STD','STA','AC']]
```

```python
data.head()
```

```python
data['DATOP'] = pd.to_datetime(data['DATOP'],errors ='coerce',format = '%Y-%m-%d')
data['DATOP'] = data['DATOP'].map(dt.datetime.timestamp)
data['STD'] = pd.to_datetime(data['STD'],errors ='coerce',format = '%Y-%m-%d %H:%M:%S')
data['STD'] = data['STD'].map(dt.datetime.timestamp)
data['STA'] = pd.to_datetime(data['STA'],errors ='coerce',format = '%Y-%m-%d %H.%M.%S')
data['STA'] = data['STA'].map(dt.datetime.timestamp)
```

```python
data.head()
```

```python
data['FLTID']=data['FLTID'].astype('category')
data['ARRSTN']=data['ARRSTN'].astype('category')
data['DEPSTN']=data['DEPSTN'].astype('category')
data['AC']=data['AC'].astype('category')

cat_columns = data.select_dtypes(['category']).columns
data[cat_columns] = data[cat_columns].apply(lambda x: x.cat.codes)
```

```python
data.head()
```

```python
data.drop('ID',axis=1,inplace=True)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('target',axis=1)
y= data['target']
y.count()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#lm = LinearRegression()
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)
```

```python
regr.fit(X_train,y_train)
```

```python
y_pred = regr.predict(X_test)
```

```python
# The coefficients
print('Coefficients: \n', regr.coef_)
```

```python
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))
```

```python
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, y_pred))
```
